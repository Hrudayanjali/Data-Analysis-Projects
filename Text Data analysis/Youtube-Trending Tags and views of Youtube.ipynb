```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
videos=pd.read_csv('C:\\Users\\sairahul\\Data analysis project\\Youtube_data_Set\\USvideos.csv',error_bad_lines=False)
```

```python
videos.head()
```

```python
#Analysing trending tags
total_tags=' '.join(videos['tags'])
```

```python
import re
```

```python
tags=re.sub('[^a-zA-Z]',' ',total_tags)
#tags
```

```python
tags=re.sub(' +',' ',tags)
#tags
```

```python
from wordcloud import WordCloud,STOPWORDS
stopwords=set(STOPWORDS)
wordcloud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(tags)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```
